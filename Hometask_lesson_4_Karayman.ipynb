### Задача 2
В ящике 5 апельсинов и 4 яблока. Наудачу выбираются 3 фрукта. Какова вероятность, что все три фрукта – апельсины?

В интернете полученный аналитически ответ 0.119. Подтверждается ли он эксперементально?


```python
import random
```

```python
# Генерируем коробку с 9 фруктами 
Box1 = ['orange']*5 + ['apple']*4
print("Взят ящик с 5 апельсинами и 4 яблоками")

# Генерируем количество 
n = 10000
print(f"Проведено {n} экспериментов")

A = pd.DataFrame({'A':list(map(lambda a: random.choice([
                            random.choice(Box1)
                            ]), range(n)
                              ))
                 })


fruit_org = ((len(A[A.A=='orange']))/3)/A.shape[0]
fruit_org
print(f"Вытащили всего апельсинов = {len(A[A.A=='orange'])}")
print(f"По итогам экспериментов максимально возможное количество комбинаций с 3 апельсинами = {int(len(A[A.A=='orange'])/3)}'")
print(f'Вероятность такой комбинации на всей выборке {fruit_org}')
```

### Не учитывая погрешность в сотых долях, экспериментально ответ подтвержден

### Задача 3
Мастер, имея 10 деталей, из которых 3 – нестандартных, проверяет детали одну за другой, пока ему не попадется стандартная. Какова вероятность, что он проверит ровно две детали?


В интернете полученный аналитически ответ 7/30 или 0.23333. Подтверждается ли он эксперементально?

```python
# Ваше решение
```

```python
Det1 = ['not_standart']*3
Det2 = ['standart']*random.choice(range(1,7)) 
#Det2 = ['standart']*7

n = 100
print(f"Проведено {n} экспериментов")


B = pd.DataFrame({'B':list(map(lambda a: random.choice([
                          random.choice(Det1),
                         random.choice(Det2)
                         ]), range(n)
                          ))
            })


det = ((len(B[B.B =='standart']))/2)/B.shape[0]
det
print(f"Вытащили всего стандартных деталей = {len(B[B.B =='standart'])}")
print(f"По итогам экспериментов максимально возможное количество комбинаций с 2 стандартными деталями = {int((len(B[B.B =='standart']))/2)}'")
print(f'Вероятность такой комбинации на всей выборке {det}')
```

### Экспериментально ответ подтверждается
