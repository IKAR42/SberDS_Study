###ML1_3: 
https://www.hackerrank.com/challenges/detect-html-links/problem?isFullScreen=true

Решение: 
import re

pattern = r'<a\ (?:.?\ )?href="((?:http)?.+?)"(?:.+?)?>\ ?([^<>]*?)?\ ?(?:<\/b>)?<\/a'
seq = re.compile(pattern)

k = int(input())
for _ in range(k):
    test = input()
    links = seq.findall(test)
    for i in links:
        print(','.join(i))
